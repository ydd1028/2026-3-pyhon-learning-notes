#### T0 extend()方法支持使用字符串扩展列表

```python
x = ['a', 'b', 'c']
x.extend(['d', 'e'])
print(x)
y = [1, 2, 3, 4, 5]
y.extend("hello")
print(y)
```

- extend()参数需要一个<mark>**可迭代对象**</mark>,然后将可迭代对象逐个插入列表

#### T1 append()

```python
s = [1, 2, 3]
s.append([4, 5, 6])
print(s)
```

- 代码不会报错，列表支持任何类型的对象，不过[4, 5, 6]**整体作为一个元素**添加到列表s中

#### T2 insert()插入

```python
x = [1, 2, 3, 4, 5]
x.insert(len(x), 6)
print(x)
```

#### T3 切片实现末尾插入

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
x[len(x):] = "上山打老虎"
y[len(y):] = ["上山打老虎"]
print(x)
print(y)
```

- 对于lst[a:b] = <mark>**可迭代对象**</mark>:
    - 先删除下标[a, b]的元素（左闭右开）
    - 再把可迭代对象挨个插入
- 字符串类型可迭代，"上山打老虎"将迭代成**5个字**
    - 实际上**大多数类型都是可迭代的**，不过整数就不可迭代
- 对于["上山打老虎"]，这是一个列表，只有**1个元素**，即"上山打老虎"
- x = y = [] --> x, y指向**同一个列表**

#### T4 括号字符串检测

```python
x = input("请输入括号字符串：")

islegal = True
stack = []

for c in x:
    if c in ('(', '[', '{'):
        stack.append(c)
    else:
        #1.处理右括号多余情况
        if stack == []:    
            islegal = False
            break

        #2.处理左右括号数量相等但不匹配情况
        if c == ')':
            d = '('
        elif c == ']':
            d = '['
        elif c == '}':
            d = '{'
        if d != stack.pop():
            islegal = False
            break

#3.处理左括号多余情况
if stack != []:
    islegal = False

if islegal:
    print("合法(*^_^*)")
else:
    print("非法(⊙﹏⊙)")
```

- `if lst[i] == '(' or '[' or '{':` 这样的表达式**永远是True!**
    - 等价于 `if (lst[i] == '(') or (']') or ('{'):`
    - 而']' 和 '}'是**非空字符串，永远是True!**
<br>

- 此题考虑3种情况：
    - 左右括号数量相等但不匹配
    - 左括号多余
    - 右括号多余
<br>

- <mark>**关于栈:**</mark>
    - 栈是一种**逻辑概念**，一种**规则**
    - *LIFO: last in first out*
    - push压栈， pop出栈
    - 应用：浏览器的后退按键、软件的撤销按钮

#### T5 栈写法再练：给定字符串s,若同一字母大小写相邻，二者一起删去

```python
s = input("请输入字符串:")
stack = []

for c in s:
    if stack and c.lower() == stack[-1].lower() and c != stack[-1]:
        stack.pop()
    else:
        stack.append(c)

print("".join(stack))
```

- 判断是否是同一字母大小写：
    - 统一转成小写/大写字母比较，但注意排除是同一个字母
    - 比较Unicode: `if(abs(ord(c) - ord(stack[-1])) == 32):`
<br>

- **列表以字符串形式输出：`print("".join(s))`**
