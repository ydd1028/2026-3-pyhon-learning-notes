#### T4 若nums[]中两数字的和等于target,输出两数字对应下标

````python
nums = []
index = []

temp = input("请输入数字，输入stop时停止")
while temp != "stop":
    nums.append(int(temp))
    temp = input("请输入数字，输入stop时停止")
    
target = int(input("请输入target:"))

for i in range(len(nums) - 1):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            index.append([i, j])
print(index)       
````

- input()返回的是**字符串类型**!
    - "2" + "7" --> "27" ,永远不等于"9"!
    - 由于后续要做数字运算，应使用**强制类型转换**转成整数类型
- 关于append()和extend
    - append(x)将x整体视为**一个元素**，添加至列表末尾
        - append( [i,j] ) --> [ [i, j] ]
    - extend(x)将可迭代对象x中的**每个元素**逐个添加至列表末尾
        - extend( [i, j ]) --> [i, j]

#### T5 上一题nums[]为100个随机数，范围是[1, 100], target由用户输入

````python
import random 
nums = []
for i in range(100):
    x = random.randint(1, 100)
    nums.append(x)

target = int(input("请输入target:"))

index = []
for i in range(100):
    for j in range(i + 1, 100):
        if nums[i] + nums[j] == target:
            index.append([i, j])

if index:
    print(index)
else:
    print("没匹配上哟(⊙﹏⊙)")
````
